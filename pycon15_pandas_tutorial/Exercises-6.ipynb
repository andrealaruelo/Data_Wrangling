### Challenge: first combine these sales together into a single dataframe, then compute how much money consumers spent on each book in each currency.

```python
sales2 = sales2.dropna(how='all')
sales2
```

```python
#Use regular exressions to get the currencies from the column Title in  sales2 
col = sales2.Title.str.extract(r'\((.*?)\)', expand=False)
col
```

```python
sales2['currency'] = col.fillna(method='bfill')
```

```python
sales1
```

```python
sales1 = sales1.rename(columns={'Book title': 'title', 'Number sold': 'sold', 'Sales price': 'price', 'Royalty paid': 'royalty'})
```

```python
sales1
```

```python
sales2 = sales2.rename(columns={'Title': 'title', 'Units sold': 'sold', 'List price': 'price', 'Royalty': 'royalty'})
```

```python
#sales = pd.merge(sales1,sales2, on=['title'], how='outer')
sales = pd.concat([sales1,sales2])
```

```python
sales
```

```python
# Fill the currency that it is still null (from sales1) 
col = sales.currency
col = col.fillna('USD')

#Assign the currency col to the df sales
sales.currency = col
```

```python
#Fill NaN
sales = sales.fillna(0)
```

```python
#There are titles of the same book that are slightly different 
colTitle = sales.title.str.replace(r'The Bricklayer.*', 'The Bricklayer\'s Bible')
colTitle = sales.title.str.replace('Pining for the Fisheries of Yore', 'Pining For The Fisheries of Yore')

#Assign the title col to the df sales
sales['title'] = colTitle
```

```python
#Remove non-relevant rows

#Suppress a particular UserWarning
import warnings
warnings.filterwarnings("ignore", 'This pattern has match groups')

sales = sales[sales.title.str.contains(r'\((.*?)\)') == False]
sales = sales[sales.title.str.contains('E-Book Reader') == False]
sales
```

```python
#Calculate the money spent
spent = sales.sold*sales.price 
sales['spent'] = spent
```

```python
#Get the money spent for each book in each currency
sales.groupby(['title', 'currency'])[['spent']].sum()
```
